```python
# import necessary modules and libraries

import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
# get dosimetry data

folder_path = '../dosimetry_data/'
dosage = 'Book2(150kV'
dists = ['_16cm', '_15cm', '_7cm', '_4cm']
tin_filter = ['_nosn)', '_sn)']
dosage_dfs = []

for dist in dists:
    for flt in tin_filter:
        filepath = Path(folder_path + dosage + dist + flt + '.csv')
        dosage_dfs.append(pd.read_csv(filepath))
```
